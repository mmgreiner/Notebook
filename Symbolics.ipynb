# Math.NET Symbolics

This is to test some symbolic algebra formulas.

```
#r "nuget: MathNet.Symbolics"

open System
open System.Numerics
open MathNet.Numerics
open MathNet.Symbolics
```

You can simply use it to reduce algebraic formulars.

```
open Operators
let x = symbol "x"
let y = symbol "y"
let a = symbol "a"
let b = symbol "b"
let c = symbol "c"
let d = symbol "d"
```

```
let aa = a + a
let kompliziert = (a/b/(c*a))*(c*d/a)/d 
Infix.format aa
```

```
kompliziert |> Infix.format 
```

```
kompliziert |> LaTeX.format
```

It is also possible to evaluate expressions.

```
let f0 = Real 2.0
let symbols = dict [ "a", FloatingPoint.Real 2.0; "b", FloatingPoint.Real 3.0 ]
printfn "symbols = %A" symbols
Evaluate.evaluate symbols (1/(a*b))  // Returns Real 0.1666666667 (as float)
```
